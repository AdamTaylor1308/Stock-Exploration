```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import numpy as np
import yfinGenHelp as ygh
import importlib
# importlib.reload(ygh)
import datetime as datetime

# ---------- CONFIG ----------

x = datetime.datetime.now().strftime('%m%d%y')

CSV_PATH = f'/Users/adamtaylor/Documents/Personal/Finance/StockDB/yfinGens/Data1.3/stock_metrics_{x}.csv'

TICKERS = [
    'AAL', 'AAPL', 'ABBV', 'ABNB', 'ABT', 'ACN', 'ADI', 'ADP', 'ADSK', 'AFRM', 'AMAT',
    'AMD', 'AMGN', 'AMKR', 'AMZN', 'ANET', 'APP', 'ARM', 'ASML', 'AVGO', 'AXP',
    'BAC', 'BE', 'BIDU', 'BIIB', 'BILI', 'BKNG', 'BLDP', 'BMY', 'BRK-B', 'BX', 'C', 'CAR',
    'CAT', 'CB', 'CDNS', 'CELH', 'CHKP', 'CHPT', 'CHTR', 'CI', 'CL', 'CMCSA', 'CME', 'CMG',
    'COP', 'COST', 'CRM', 'CRSP', 'CRWD', 'CSCO', 'CSGP', 'CVX', 'DASH', 'DE', 'DG', 'DIS',
    'DKNG', 'DLTR', 'DPZ', 'DUK', 'EA', 'ENPH', 'ETSY', 'EVGO', 'FISV', 'FIVE', 'FLEX',
    'FOXA', 'FSLR', 'FTNT', 'FVRR', 'GD', 'GE', 'GILD', 'GGG', 'GIS', 'GOOG', 'GOOGL', 'GPN', 'GRAB',
    'GS', 'HD', 'HON', 'HSY', 'IBM', 'ILMN', 'INTC', 'INTU', 'ISRG', 'JBL', 'JNJ', 'JPM',
    'KLAC', 'KMB', 'KO', 'KR', 'LLY', 'LMT', 'LOW', 'LRCX', 'LULU', 'LYFT', 'MA', 'MAR',
    'MCD', 'MCHP', 'MDB', 'MDLZ', 'MDT', 'MELI', 'META', 'MMC', 'MO', 'MPWR', 'MRK', 'MRNA',
    'MS', 'MSFT', 'MTCH', 'MU', 'NEE', 'NET', 'NFLX', 'NOW', 'NTNX', 'NVDA', 'NXPI',
    'OKTA', 'ON', 'ORCL', 'PANW', 'PATH', 'PAYX', 'PEP', 'PFE', 'PG', 'PGR', 'PLTR',
    'PLUG', 'PM', 'PTON', 'PZZA', 'QCOM', 'QQQ', 'QRVO', 'RBLX', 'RCL', 'REGN', 'ROKU', 'ROST',
    'RTX', 'SBUX', 'SCHW', 'SE', 'SEDG', 'SHOP', 'SIRI', 'SJM', 'SNPS', 'SOFI', 'SOXL', 'SPGI',
    'SPY', 'STX', 'SWKS', 'SYK', 'T', 'TEAM', 'TGT', 'TMO', 'TMUS', 'TQQQ', 'TRMB', 'TSLA',
    'TSM', 'TTWO', 'TXN', 'U', 'UBER', 'ULTA', 'UNH', 'UNP', 'UPS', 'UPST', 'URI', 'V',
    'VFC', 'VRTX', 'WDAY', 'WDC', 'WFC', 'WING', 'WMT', 'WW', 'XOM', 'YUM', 'ZG', 'ZS',
    'ZTS'
]

STOCK_LIST = TICKERS[:]

MUTUAL_LIST = ['VTSAX', 'VFIAX', 'FXAIX', 'VTIAX', 'VBTLX', 'VIMAX', 'VTSMX', 'VTSPX', 'VIMSX', 
               'VSMPX', 'FCNTX', 'AGTHX', 'TRBCX', 'SWPPX', 'PTTRX', 'FDRXX', 'VMMXX', 'VFINX', 
               'VIGAX', 'VVIAX', 'PRBLX', 'FOCPX', 'VTBIX', 'VINIX', 'CGFFX', 'GFACX', 'VTHRX', 
               'AEPGX', 'VSIAX', 'VMVAX', 'VIGRX', 'PRJTX', 'TEMIX', 'TROSX', 'DODBX', 'DODSX', 
               'RWMFX', 'RGAGX', 'RGHTX', 'SGOVX', 'MNYXX', 'FRCXX', 'PREMX', 'WIGTX', 'FSMDX', 
               'FSCSX', 'FSPGX', 'FBALX', 'FBGRX', 'FLCOX', 'FMAGX', 'HEDTX', 'OAKBX', 'MFEIX', 
               'JGMIX', 'PCOMX', 'WSHFX', 'GLACX', 'CEFWX', 'AEDFX', 'IEUIX', 'TLINX', 'VBMFX', 
               'VTBSX', 'VIPSX', 'WFZIX', 'MWTRX', 'PCRAX', 'JATTX', 'PIGIX', 'PTTRX', 'MCLDX', 
               'ABNDX', 'TOTL', 'BND', 'AGG', 'LQD', 'JNK', 'HYG', 'SHY', 'IEF', 'TLT', 'BSV', 
               'VCSH', 'VGIT', 'MBB', 'BNDX', 'EMB', 'VWOB', 'EFA', 'VEA', 'IEFA', 'IEMG', 'VWO', 
               'IXUS', 'VXUS', 'VUE', 'VOO', 'IVV', 'SPY', 'VTI', 'QQQ', 'VUG', 'VTV', 'BND', 
               'AGG', 'GLD', 'IWF', 'IEMG', 'VXUS', 'VGT', 'VWO', 'VIG', 'IJH', 'SPYM', 'XLK', 
               'VO', 'IJR', 'ITOT', 'RSP', 'BNDX', 'IWM', 'SCHD', 'QQQM', 'VB', 'EFA', 'VYM', 
               'IWD', 'IVW', 'IAU', 'SGOV', 'SCHX', 'VCIT']

ETF_LIST = ['VOO', 'IVV', 'SPY', 'VTI', 'QQQ', 'VUG', 'VEA', 'IEFA', 'VTV', 'BND', 'GLD', 'AGG', 
            'IWF', 'IEMG', 'VXUS', 'VGT', 'VWO', 'VIG', 'IJH', 'SPYM', 'XLK', 'VO', 'IJR', 'ITOT', 
            'RSP', 'BNDX', 'IBIT', 'IWM', 'SCHD', 'QQQM', 'EFA', 'VB', 'VYM', 'IWD', 'IVW', 'IAU', 
            'SGOV', 'SCHX', 'VCIT', 'VT', 'SCHF', 'SCHG', 'VEU', 'XLF', 'IXUS', 'TLT', 'QUAL', 'IVE', 
            'VV', 'IWR', 'IWB', 'SPYG', 'IEF', 'BIL', 'VTEB', 'BSV', 'MUB', 'DIA', 'JEPI', 'XLV', 
            'VCSH', 'DFAC', 'MBB', 'SCHB', 'SMH', 'VGIT', 'VONG', 'DGRO', 'JPST', 'VNQ', 'IUSB', 
            'GOVT', 'LQD', 'MGK', 'SPDW', 'JEPQ', 'VBR', 'TQQQ', 'SPYV', 'DYNF', 'SLV', 'OEF', 'VGK', 
            'XLE', 'EFV', 'BIV', 'XLC', 'IUSG', 'CGDV', 'VGSH', 'USHY', 'JAAA', 'VXF', 'XLI', 'ACWI', 
            'IUSV', 'GLDM', 'GDX', 'SHY', 'IDEV' ]

EXCHANGE = 'NASDAQ'
MIN_PRICE = 5
SOURCE = 'yahoo'
FETCH_METADATA = True

# ---------- HELPER: VECTOR ASSET TYPE ----------
asset_type_map = (
    {t: "STOCK" for t in STOCK_LIST} |
    {t: "MUTUAL_FUND" for t in MUTUAL_LIST} |
    {t: "ETF" for t in ETF_LIST}
)

def assign_asset_type_series(ticker_series: pd.Series) -> pd.Series:
    return ticker_series.map(asset_type_map).fillna("UNKNOWN")


# ---------- METRICS HELPERS ----------

def compute_rsi_prices(prices: pd.DataFrame, period: int = 14) -> pd.DataFrame:
    """Vectorized RSI for all tickers at once."""
    delta = prices.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_metrics(series: pd.Series) -> pd.DataFrame:
    """Per-ticker metrics; series is the Close price history for one ticker."""
    df = pd.DataFrame({'close': series})
    df['daily_return'] = df['close'].pct_change()

    df['return_1m'] = df['close'].pct_change(21)
    df['return_3m'] = df['close'].pct_change(63)
    df['return_12m'] = df['close'].pct_change(252)

    df['ma_50'] = df['close'].rolling(50).mean()
    df['ma_50_30d_ago'] = df['ma_50'].shift(30)
    df['ma_200'] = df['close'].rolling(200).mean()

    df['rsi_14'] = compute_rsi_prices(df[['close']], 14)['close']  # reuse vector logic

    df['vol_3m'] = df['daily_return'].rolling(63).std()

    df['downside_return'] = df['daily_return'].where(df['daily_return'] < 0, 0)
    df['downside_vol_3m'] = df['downside_return'].rolling(63).std()

    df['sharpe_1y'] = (
        df['daily_return'].rolling(252).mean() /
        df['daily_return'].rolling(252).std()
    ) * np.sqrt(252)

    return df

def compute_beta_and_idio_for_ticker(stock_returns: pd.Series,
                                     spy_returns,
                                     window: int = 252,
                                     min_obs: int = 60):
    """
    Compute 1-year beta and idiosyncratic volatility for a single ticker
    using overlapping dates with SPY.
    """

    # --- Force SPY to be a clean Series ---
    if isinstance(spy_returns, pd.DataFrame):
        # use the first column no matter what it's called
        market_series = spy_returns.iloc[:, 0].copy()
    else:
        market_series = spy_returns.copy()

    market_series.name = "market"

    # --- Ensure stock is a clean Series ---
    stock_series = stock_returns.copy()
    stock_series.name = "stock"

    # --- Align on dates ---
    aligned = pd.concat([stock_series, market_series], axis=1, join="inner").dropna()

    # Not enough history
    if len(aligned) < max(window, min_obs):
        return np.nan, np.nan

    recent = aligned.tail(window)

    # Extract arrays
    s = recent["stock"].to_numpy()
    m = recent["market"].to_numpy()

    # Compute centered versions
    s_mean = s.mean()
    m_mean = m.mean()

    s_c = s - s_mean
    m_c = m - m_mean

    var_m = (m_c ** 2).mean()
    if var_m == 0:
        return np.nan, np.nan

    # Beta
    cov_sm = (s_c * m_c).mean()
    beta = cov_sm / var_m

    # Alpha
    alpha = s_mean - beta * m_mean

    # Residuals for idiosyncratic volatility
    pred = alpha + beta * m
    resid = s - pred
    idio_vol = resid.std(ddof=1)

    return beta, idio_vol

def compute_signal_vectorized(df: pd.DataFrame, vol_cutoff: float) -> pd.Series:
    cond = (
        (df['close'] > df['ma_50']) &
        (df['close'] > df['ma_200']) &
        (df['ma_50'] > df['ma_200']) &
        (df['ma_50_30d_ago'] < df['ma_50']) &
        (df['r3m'] >= 0) &
        (df['r12m'] >= df['r3m']) &
        (df['rsi'] >= 53) &
        (df['vol_3m'] <= vol_cutoff)
    )
    return np.where(cond, 'Y', 'N')


# Download PRICES (CHUNKED, SHORTER PERIOD) ----------

print("Downloading historical prices in batch...")

frames = []
failed_all = []

for chunk in ygh.chunk_list(TICKERS, 50):
    print(f"Downloading chunk: {chunk}")
    try:
        df_chunk = yf.download(
            chunk,
            period="400d",
            auto_adjust=False,
            timeout=30
        )['Close']
        frames.append(df_chunk)
    except Exception as e:
        print("Batch failed:", e)
        failed_all.extend(chunk)

if not frames:
    raise RuntimeError("No price data downloaded at all – check connection or ticker list.")

price_data = pd.concat(frames, axis=1)

# Retry any failed tickers individually (change period?)
if failed_all:
    print("Retrying failed tickers individually:", failed_all)
    for t in failed_all:
        try:
            df_single = yf.download(t, period="400d", auto_adjust=False, timeout=30)['Close']
            price_data[t] = df_single
        except Exception as e:
            print(f"Final fail for {t}: {e}")

# Ensure columns align with TICKERS we requested (drop extras if any)
price_data = price_data.loc[:, [c for c in TICKERS if c in price_data.columns]]

# SPY Benchmark
print("Downloading SPY benchmark...")
spy_data = yf.download("SPY", period="400d", auto_adjust=False)['Close']
spy_returns = spy_data.pct_change()

# Processing each ticker
rows = []

for ticker in price_data.columns:
    series = price_data[ticker].dropna()

    if len(series) == 0:
        continue

    latest_price = series.iloc[-1]
    if latest_price < MIN_PRICE:
        continue

    hist = calculate_metrics(series)
    latest = hist.iloc[-1]

    # Metadata: optional for speed
    company_name = ticker
    sector = 'Unknown'
    market_cap = None

    if FETCH_METADATA:
        try:
            info = yf.Ticker(ticker).info or {}
            company_name = info.get('longName', ticker)
            sector = info.get('sector', 'Unknown')
            market_cap = info.get('marketCap', None)
        except Exception as e:
            print(f"Metadata fetch failed for {ticker}: {e}")

    stock_ret_1y = hist['daily_return']   

    beta, idio_vol_1y = compute_beta_and_idio_for_ticker(
        stock_ret_1y,
        spy_returns,
        window=252,
        min_obs=60   # adjust?
    )

    downside_vol_3m = latest['downside_vol_3m']


    row = {
        'ticker': ticker,
        'exchange': EXCHANGE,
        'company_name': company_name,
        'close': latest['close'],
        'daily_return': latest['daily_return'],
        'rsi': latest['rsi_14'],
        'r1m': latest['return_1m'],
        'r3m': latest['return_3m'],
        'r12m': latest['return_12m'],
        'ma_50': latest['ma_50'],
        'ma_50_30d_ago': latest['ma_50_30d_ago'],
        'ma_200': latest['ma_200'],
        'vol_3m': latest['vol_3m'],
        'downside_vol_3m': downside_vol_3m,
        'idiosyncratic_vol_1y': idio_vol_1y,
        'sharpe_1y': latest['sharpe_1y'],
        'beta_1y': beta,
        'market_cap': market_cap,
        'sector': sector,
        'eval_date': latest.name.strftime('%Y-%m-%d'),
        'source': SOURCE
    }

    rows.append(row)

df_out = pd.DataFrame(rows)

if df_out.empty:
    raise RuntimeError("No valid rows produced – check MIN_PRICE, data availability, etc.")

# Asset Type
df_out['asset_type'] = assign_asset_type_series(df_out['ticker'])

# Z-Scores & Safety Score

vol_mean = df_out['vol_3m'].mean()
vol_std = df_out['vol_3m'].std()
df_out['vol_z'] = (vol_mean - df_out['vol_3m']) / vol_std  

# Cross-sectional z-scores (ddof=0 to match your original)
df_out['sharpe_1y_z'] = (df_out['sharpe_1y'] - df_out['sharpe_1y'].mean()) / df_out['sharpe_1y'].std(ddof=0)
df_out['downside_vol_z'] = (df_out['downside_vol_3m'] - df_out['downside_vol_3m'].mean()) / df_out['downside_vol_3m'].std(ddof=0)
df_out['idiosyncratic_vol_z'] = (df_out['idiosyncratic_vol_1y'] - df_out['idiosyncratic_vol_1y'].mean()) / df_out['idiosyncratic_vol_1y'].std(ddof=0)

# Safety Score 
df_out['safety_score'] = (
    -0.1701 * df_out['sharpe_1y_z'] +
     0.1956 * df_out['vol_z'] +
    -0.6432 * df_out['downside_vol_z'] +
     0.3813 * df_out['idiosyncratic_vol_z']
)

# Signal

vol_cutoff = df_out['vol_3m'].quantile(0.80)
df_out['signal'] = compute_signal_vectorized(df_out, vol_cutoff)

# Save to CSV

df_out.to_csv(CSV_PATH, index=False)
print(f"\nSaved {len(df_out)} tickers to {CSV_PATH}")
```
